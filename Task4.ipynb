1: Mobile Recharge Validation System



```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    # Check minimum amount
    if amount < 50:
        return "Recharge amount must be at least ₹50."

    # Check if plan is valid
    if amount not in valid_plans:
        return "Invalid plan selected. Available plans: 199, 299, 399, 599."

    return "Recharge Successful!"

# Retry system using while loop
while True:
    try:
        recharge_amount = int(input("Enter recharge amount: "))
        result = validate_recharge(recharge_amount)
        print(result)

        if result == "Recharge Successful!":
            break
    except ValueError:
        print("Please enter a valid numeric amount.")
```

2: Inventory Reorder Alert System



```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert (Stock: {stock})")
        else:
            print(f"{product}: Stock OK (Stock: {stock})")

# Sample inventory dictionary
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18
}

check_inventory(inventory)
```

3: Student Result Processing System



```python
def process_results(marks_list):
    total = 0

    # Calculate total using loop
    for mark in marks_list:
        total += mark

    average = total / len(marks_list)

    if average >= 50:
        return f"Average: {average:.2f} - Pass"
    else:
        return f"Average: {average:.2f} - Fail"

# Example usage
marks = [60, 45, 70, 55]
print(process_results(marks))
```

4: Cab Fare Estimator with Retry Option



```python
def calculate_fare(distance, peak_hour):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    if peak_hour:
        fare += fare * 0.25   # 25% extra

    return fare

# Retry system
while True:
    try:
        km = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ").lower()

        peak_status = True if peak == "yes" else False

        total_fare = calculate_fare(km, peak_status)
        print(f"Total Fare: ₹{total_fare:.2f}")
        break
    except ValueError:
        print("Please enter valid input.")
```

5: Employee Attendance Eligibility Checker



```python
def check_attendance(attendance_list):
    present_days = 0

    # Count present days using loop
    for day in attendance_list:
        if day.upper() == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% - Eligible"
    else:
        return f"Attendance: {percentage:.2f}% - Not Eligible"

# Example usage
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]
print(check_attendance(attendance))
```

6: Password Strength Checker



```python
def check_password(password):
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password: Minimum length should be 8."

    has_digit = False
    has_special = False

    # Loop through each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if not has_digit:
        return "Weak Password: Must contain at least one digit."

    if not has_special:
        return "Weak Password: Must contain at least one special character (@#$)."

    return "Strong Password!"

# Example usage
pwd = input("Enter password: ")
print(check_password(pwd))
```
