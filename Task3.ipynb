1: Employee Performance Bonus Eligibility


```python
# Given dictionary of employee performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find highest score
highest_score = max(employees.values())

# Step 2: Identify employees with highest score (handle ties)
top_performers = [name for name, score in employees.items() if score == highest_score]

# Step 3: Display result
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

2: Search Query Keyword Analysis


```python
import string

# Input search query
query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Step 3: Split into words
words = query.split()

# Step 4: Count word frequency
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1

# Step 5: Filter words searched more than once
frequent_words = {word: count for word, count in word_count.items() if count > 1}

print(frequent_words)
```

3: Sensor Data Validation


```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Step 1: Identify even readings and store (index, value)
valid_readings = [(index, value) for index, value in enumerate(sensor_readings) if value % 2 == 0]

# Step 2: Display results
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

4: Email Domain Usage Analysis


```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Step 1: Count domain occurrences
domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Step 2: Calculate percentage usage
total_emails = len(emails)

print("Email Domain Usage Percentage:")
for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

5: Sales Spike Detection


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate average sales
average_sales = sum(sales) / len(sales)

# Step 2: Detect spikes (more than 30% above average)
threshold = average_sales * 1.30

print("Sales Spikes Detected:")
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Step 1: Count occurrences
id_count = {}

for user in user_ids:
    id_count[user] = id_count.get(user, 0) + 1

# Step 2: Identify duplicates
print("Duplicate User IDs:")
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
