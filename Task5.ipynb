1: Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes_list):
    total_likes = 0

    # Calculate total likes using loop
    for likes in likes_list:
        total_likes += likes

    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"

    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

# Example usage
likes = [200, 300, 150, 400]
analyze_engagement(likes)
```

3: Agriculture – Rainfall Adequacy Checker

```python
def rainfall_analysis(rainfall_list, required_level):
    total = 0

    for rain in rainfall_list:
        total += rain

    average = total / len(rainfall_list)

    status = "Adequate Rainfall" if average >= required_level else "Inadequate Rainfall"

    print(f"Average Rainfall: {average}")
    print(f"Rainfall Status: {status}")

# Example usage
rainfall_data = [60, 75, 80, 70]
rainfall_analysis(rainfall_data, 70)
```

4: Social Media – Duplicate Account Detection

```python
def detect_duplicates(usernames):
    unique_users = set(usernames)

    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

# Example usage
users = ["user1", "user2", "user3", "user1"]
detect_duplicates(users)
```

6: Agriculture – Premium Crop Price Filter

```python
def filter_premium_crops(price_list):
    premium = []

    for price in price_list:
        if price > 2000:
            premium.append(price)

    print(f"Premium Crops: {premium}")

# Example usage
crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

7: System Monitoring – Application Health Checker

```python
def check_system_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")

# Example usage
check_system_health(7)
```

9: E-Learning – Student Attendance Eligibility System

```python
def check_exam_eligibility(attendance_list):
    present = 0

    for day in attendance_list:
        if day.upper() == "P":
            present += 1

    percentage = (present / len(attendance_list)) * 100

    status = "Eligible" if percentage >= 75 else "Not Eligible"

    print(f"Attendance Percentage: {percentage}")
    print(f"Exam Eligibility: {status}")

# Example usage
attendance = ["P", "P", "A", "P", "P"]
check_exam_eligibility(attendance)
```

10: Smart Electricity Bill Analyzer

```python
def calculate_electricity_bill(units):
    bill = 0

    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage

# Example usage
total_bill, usage_status = calculate_electricity_bill(250)

print(f"Total Bill: ₹{total_bill}")
print(f"Usage Classification: {usage_status}")
```
